# 2)Treat missing values in CPC, CTR and CPM using the formula given :

FORMULAS GIVEN :

         CTR = Clicks/Impresssions * 100
         CPM = Spend/Impressions * 1000
         CPC = Spend/Clicks 
    

```python
def ctr(x):
    Clicks=digital.Clicks
    Impressions=digital.Impressions
    ctr= Clicks/Impressions * 100
    return ctr
```

```python
def cpm(x):
    Impressions=digital.Impressions
    Spend=digital.Spend
    cpm=Spend/Impressions*1000
    return cpm
```

```python
digital['CPM']= digital[['CPM']].apply(lambda x: cpm(x))
```

```python
def cpc(x):
    Clicks=digital.Clicks
    Spend=digital.Spend
    cpc=Spend/Clicks
    return cpc
```

# 3) Check if there are any outliers. Do you think treating outliers is necessary for K-Means clustering? Based on your judgement decide whether to treat outliers and if yes, which method to employ. (As an analyst your judgement may be different from another analyst).

       Outlier detection is an important data analysis task and treating the outliers from clusters can improve the 
       clustering accuracy.So we have to treat outlier before applying k means algorithm

```python
digital.info()
```

```python
dig_num=[]
dig_cat=[]
for i in digital.columns:
    if digital[i].dtypes=='object':
        dig_cat.append(i)
    else:
        dig_num.append(i)
print('dig_num:',dig_num)
print('dig_cat:',dig_cat)
```

```python
digital[dig_num].describe()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=digital)
plt.xticks(rotation=90)
plt.show();
```

            Except Ad length and Ad width, others having outliers.

```python
def remove_outliers(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lr=Q1-(1.5*IQR)
    ur=Q3+(1.5*IQR)
    return lr,ur
```

```python
for i in dig_num:
    LL,UL=remove_outliers(digital[i])
    digital[i]=np.where(digital[i]>UL,UL,digital[i])
    digital[i]=np.where(digital[i]<LL,LL,digital[i])
    
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=digital)
plt.xticks(rotation=90)
plt.show();
```

#        Thus using BOXPLOT Method we can treat outliers to get high accuracy while performing K-Clustering Algorithm.
